```python
df = pd.read_csv("./ml-25m/ratings.csv")
df.head()
```

```python
# Training data, Test data slicing
train_df = df[(df['timestamp'] >= 1104505203) & (df['timestamp'] <= 1230735592)]
test_df = df[df['timestamp'] >= 1230735600]
```
